```python
print("☕Welcome to our Coffee shop!☕")
print("---------------------------")

menu={
    "tea" : 25,
    "samosa":15,
    "veg puffs": 15,
    "egg puffs": 20,
    "paneer puffs": 25,
    "mushroom puffs": 30,
    "coffee": 25,
    "cool drinks": 50,
}

#print(f"tea: Rs25\n samosa: Rs15\n veg puffs: Rs15\n egg puffs: Rs20\n paneer puffs: Rs25\n mushroom puffs: Rs30\n coffee: Rs25\n cool drinks: Rs50")

cart={}   #cart dictionary
order_total=0


while True:
  item=input("Enter item name: ")

  if item in menu:
    qty=int(input("Enter quantity: "))

    #Add cart
    if item in cart:
      cart[item]+=qty
    else:
      cart[item]=qty

    print(f"{qty} {item} added to card")
  else:
    print("Item not available")

  another_order=input("Do you want ot add another item? (yes/no):")

  if another_order != "yes":
    break

#Bill SECTION
print("\n------BILL-------")

for item,qty in cart.items():
  price=menu[item]
  item_total=price*qty
  order_total+=item_total
  print(f"{item} * {qty} = Rs {item_total}")

#GST Calculation

gst_percent=5
gst_amount=(order_total * gst_percent)/100
final_amount=order_total+gst_amount


print(f"subtotal: Rs {order_total:.2f}")
print(f"GST(5%): Rs {gst_amount:.2f}")
print("----------------------------")
print(f"\n Total Amount to pay is RS{final_amount:.2f}")
print("-----------------------------")
print("Thank you!visit again☕")
```
