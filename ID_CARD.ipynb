```python
from PIL import Image,ImageDraw,ImageFont
import qrcode
```

```python
from IPython.display import display
```

```python
import random
import os
import datetime
```

```python
#card size.

card=Image.new("RGB",(500,350),"white")
draw=ImageDraw.Draw(card)
```

```python
#Header

draw.rectangle([(0,0),(400,50)],fill=(0,102,204)) #fill rectangle with color
draw.text((120,15),"ID CARD", fill="white")
```

```python
d=datetime.datetime.now()
fo=d.strftime("%d-%m-%Y\t ID CARD\t %I:%M:%S %p")
print(fo)
```

```python
try:
  font=ImageFont.truetype("arial.ttf",40)
except:
  font=ImageFont.load_default()
```

```python
na=input("Enter the name: ")
id=input("Enter the id: ")
ad=input("Enter the address:")
ph=input("Enter the phoneno: ")
com=input("Enter the company: ")
blo=input("Enter the bloodgroup: ")

text=(
    f"{'Name'.ljust(11)}       : {na}\n"
    f"{'emp_id'.ljust(11)}     : {id}\n"
    f"{'address'.ljust(11)}    : {ad}\n"
    f"{'phone'.ljust(11)}      : {ph}\n"
    f"{'Company'.ljust(11)}    : {com}\n"
    f"{'bloodgroup'.ljust(11)} : {blo}"
)
draw.text((50,50),text,fill=(0,0,0),font=font)
```

```python
qr_data=f"Name : {na}, emp_id : {id}, address  : {ad}, phoneno  :{ph}, company  :{com}, bloodgroup : {blo}"
qr=qrcode.make(qr_data)
qr=qr.resize((80,80))

card.paste(qr,(300,140))
```

```python
card.save("id_card.png")
display(card)
```
