```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data=pd.read_csv(r"/content/Student Study data.csv")
data.head()
```

```python
data=data.dropna()
```

```python
#cleaning the data
#.info() basic information about the dataframe.
data.info()
```

```python
#drop duplicate values.
data.drop_duplicates()
```

```python
#null values
data['Marital Status'].fillna("0.0", inplace=True)
data
data['study hour'].fillna("0.0", inplace=True)
data
```

```python
#statistical summary about the dataset.
data.describe()
```

```python
data.isnull().sum()
```

```python
#rename column
data.rename(columns={'Marital Status':'Marital_Status','Your gender?':'Gender','study hour':'Study Hour','your sleep hour?':'Sleep Hour','weekend':'Week End','what is your age?':'Age','what is your cgpa':'CPGA'},inplace=True)
```

```python
data.columns
```

```python
#for marital status lets assume that 0 is single and 1 is married &gender 1 is male and 2 is female.
s={0:'Single',1:'Married'}
data['Marital_Status']=data['Marital_Status']
g={1:'Male',2:'Female'}
data['Gender']=data['Gender']
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='Day',y='Study Hour',data=data)
plt.title('Title')
plt.xlabel('Sleep Hour')
plt.ylabel('Study Hour')
plt.show()
```
